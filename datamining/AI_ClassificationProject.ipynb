```python
# Data
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
voice = pd.read_csv('./voice_ver2.csv')
voice.head()
```

```python
voice.describe()
```

```python
voice.isnull().sum()
```

```python
print("Total number of labels: %d" %voice.shape[0])
print("Number of male: %d" %voice[voice.label == 'male'].shape[0])
print("Number of female: %d" %voice[voice.label == 'female'].shape[0])
```

```python
voice_label = pd.Categorical(voice['label']).codes
voice_label
```

```python
voice_label = voice_label.reshape(-1,1)
voice_label
```

```python
voice_feature = voice.iloc[:,:-1].values
voice_feature
```

```python
data_voice = np.hstack([voice_feature, voice_label])
data_voice
```

```python
# Preprocessing Features 표준화 시키는 과정
mean_voice_feature = voice_feature.mean(axis=0) # axis=0 열마다 평균을 구함
pre_voice_feature = voice_feature-mean_voice_feature
std_voice_feature = voice_feature.std(axis=0) # 열마다 sd을 구함
pre_voice_feature = pre_voice_feature/std_voice_feature # 표준화 시킴
```

```python
pd_pre_voice_feature = pd.DataFrame(pre_voice_feature, columns=['meanfreq', 'sd', 'median', 'Q25', 'Q75', 'IQR', 'skew', 'kurt',
       'sp.ent', 'sfm', 'mode', 'centroid', 'meanfun', 'minfun', 'maxfun',
       'meandom', 'mindom', 'maxdom', 'dfrange', 'modindx'])
```

```python
pd_pre_voice_feature.describe()
```

```python
# Split_Train Dataset, Test Dataset
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(pre_voice_feature, voice_label, test_size=0.2, random_state=1)
```

```python
print("Number of train data: {}".format(y_train.shape[0]))
print("Number of test data: {}".format(y_test.shape[0]))
```

```python
y_test
```

```python
# SVM
from sklearn.svm import SVC
from sklearn import metrics
```

```python
svc=SVC(C=0.3, gamma=0.05, kernel='linear') #Default hyperparameters
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print('Accuracy Score:')
print(np.round(metrics.accuracy_score(y_test,y_pred)*100,0),'%')
```

```python
# SVM_K-fold CrossValidation 데이터가 적을 때 사용
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
```

```python
svm_total=[]
svm_total_acc=[]
kf = KFold(n_splits=10, shuffle=True)
i = 1
for train_index, test_index in kf.split(pre_voice_feature):
    x_train, x_test = pre_voice_feature[train_index], pre_voice_feature[test_index]
    y_train, y_test = voice_label[train_index], voice_label[test_index]
    #print("TRAIN:", train_index, "TEST:", test_index)
    svc.fit(x_train, y_train)
    y_pred = svc.predict(x_test)
    svm_acc=metrics.accuracy_score(y_test,y_pred)
    print(svm_acc)
    svm_total_acc.append(svm_acc)
    svm_cm = confusion_matrix(y_test, y_pred)
    print('confusion matrix {}\n'.format(i), svm_cm)
    svm_total.append(svm_cm)
    i += 1
```

```python
# K-Fold SVM Accuracy
print('SVM accuracy :', np.round(np.mean(svm_total_acc)*100,1),'%')
```

```python
# Sum SVM confusion matrices
svm_confusion = np.array([[0,0],[0,0]])

for i in range(10):
    svm_confusion += svm_total[i]
svm_confusion
```

```python
conf_matrix=pd.DataFrame(svm_confusion,columns=['Predicted:0(Female)','Predicted:1(Male)'],index=['Actual:0(Female)','Actual:1(Male)'])
plt.figure(figsize = (8,6))
plt.title('SVM Confusion Matrix', fontsize = 15)
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu")
```

```python
# Perceptron
x_train, x_test, y_train, y_test = train_test_split(pre_voice_feature, voice_label, test_size=0.2, random_state=1)
```

```python
print("Number of train data: {}".format(y_train.shape[0]))
print("Number of test data: {}".format(y_test.shape[0]))
```

```python
from sklearn.linear_model import Perceptron
```

```python
# Default hyperparameters
Per=Perceptron(alpha=0.001, tol= 0.005, penalty='l2') 
Per.fit(x_train, y_train)
y_pred = Per.predict(x_test)
print('Accuracy Score:')
print(np.round(metrics.accuracy_score(y_test,y_pred)*100,1),'%')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
```

```python
Per_total=[]
Per_total_acc=[]
kf = KFold(n_splits=10, shuffle=True)
i = 1
for train_index, test_index in kf.split(pre_voice_feature):
    x_train, x_test = pre_voice_feature[train_index], pre_voice_feature[test_index]
    y_train, y_test = voice_label[train_index], voice_label[test_index]
    #print("TRAIN:", train_index, "TEST:", test_index)
    Per.fit(x_train, y_train)
    y_pred = Per.predict(x_test)
    Per_acc=metrics.accuracy_score(y_test,y_pred)
    print(Per_acc)
    Per_total_acc.append(Per_acc)
    Per_cm = confusion_matrix(y_test, y_pred)
    print('confusion matrix {}\n'.format(i), Per_cm)
    Per_total.append(Per_cm)
    i += 1
```

```python
print('Perceptron accuracy :', np.round(np.mean(Per_total_acc)*100,1),'%')
```

```python
Per_confusion = np.array([[0,0],[0,0]])

for i in range(10):
    Per_confusion += Per_total[i]
Per_confusion
```

```python
conf_matrix=pd.DataFrame(Per_confusion,columns=['Predicted:0(Female)','Predicted:1(Male)'],index=['Actual:0(Female)','Actual:1(Male)'])
plt.figure(figsize = (8,6))
plt.title('Perceptron Confusion Matrix', fontsize = 15)
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu")
```
