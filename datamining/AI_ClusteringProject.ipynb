```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing as pp
from sklearn.cluster import KMeans

sns.set()
%matplotlib inline
```

```python
iris = pd.read_csv("./Iris.csv")
iris.shape
```

```python
iris.head()
```

```python
X = iris.iloc[:,1:5].values
X.shape
```

```python
X
```

```python
temp = iris['Species'].values
temp
```

```python
y = pd.Categorical(iris['Species']).codes
y.shape
```

```python
Species = y.reshape(-1,1)
Species.shape
```

```python
Species
```

```python
data = np.hstack([X,Species])
```

```python
data
```

```python
df = pd.DataFrame(data, columns=["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm","Species"])
```

```python
df
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(15,15))
plt.title('Correlation between data variables', fontsize = 20)
heatmap = sns.heatmap(correlation_matrix ,annot=True, fmt='.2f',linewidth=1,cmap='Blues')
heatmap.set_ylim(0,5)
```

```python
# Data Select
val_SepalLengthCm = df['SepalLengthCm'].values
val_PetalLengthCm = df['PetalLengthCm'].values
val_PetalWidthCm = df['PetalWidthCm'].values
```

```python
val_SepalLengthCm = val_SepalLengthCm.reshape(-1,1)
val_PetalLengthCm = val_PetalLengthCm.reshape(-1,1)
val_PetalWidthCm = val_PetalWidthCm.reshape(-1,1)
```

```python
data_values = np.hstack([val_SepalLengthCm, val_PetalLengthCm, val_PetalWidthCm])
```

```python
data_values
```

```python
# Clustering_Algorithm
from sklearn.cluster import KMeans

cost = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state = 0).fit(data_values)
    cost.append(abs(kmeans.score(data_values)))

plt.figure(figsize = (10,7))
plt.title('The K-means Clustering cost', fontsize = 15)
plt.plot(range(1,11), cost)
plt.xticks([2,4,6,8,10])
plt.xlim([0.5, 10.5])
plt.xlabel('number of cluster', fontsize = 12)
plt.ylabel('cost', fontsize = 12)
plt.grid(alpha = 0.3)
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters = 3, init='k-means++', random_state = 0)
y_kmeans= kmeansmodel.fit_predict(data_values)
```

```python
real_values = df["Species"].astype('int')
```

```python
print("Result of kmeans : ")
real_unique_class , real_unique_class_counts = np.unique(y_kmeans, return_counts=True)
for x , y in zip(real_unique_class, real_unique_class_counts):
    print("The number of observations assigned cluster",x,"is",y)
```

```python
# Accuracy_Calculation
temp = iris['Species'].values
real_species = temp.reshape(-1,1)
real_species
```

```python
# True Species Class와 y_kmeans Cluster가 다를 경우 조정
kmeans_species = []
for sp in y_kmeans:
    if sp == 0:
        sp = 'Iris-setosa'
        kmeans_species.append(sp)
    elif sp == 1:
        sp = 'Iris-versicolor'
        kmeans_species.append(sp)
    else:
        sp = 'Iris-virginica'
        kmeans_species.append(sp)
```

```python
kmeans_species
```

```python
real_species.shape
```

```python
kmeans_species = np.array(kmeans_species)
kmeans_species = kmeans_species.reshape(-1,1)
kmeans_species.shape
```

```python
i = 0 # predict = Real
j = 0 # total

for real, predict in zip(real_species, kmeans_species):
    print(real,predict)
    if real == predict:
        i += 1
    j += 1
    
print('i :',i,', j :',j)
```

```python
accuracy = np.round((i/j)*100,1)
accuracy
```

```python
i = 0 # setosa, real = prediction 개수
setosa_num = 0
j = 0 # versicolor, real = prediction 개수
versicolor_num = 0
k = 0 # virginica, real = prediction 개수
virginica_num = 0

for real, predict in zip(real_species, kmeans_species):
    if real == 'Iris-setosa':
        setosa_num += 1
        if real == predict:
            i += 1
    elif real == 'Iris-versicolor':
        versicolor_num += 1
        if real == predict:
            j += 1
    else:
        virginica_num += 1
        if real == predict:
            k += 1
print(i,j,k)
```

```python
# accuracy
ac_setosa = np.round((i/setosa_num)*100, 1)
print('K-means setosa accuracy :{}'.format(ac_setosa), '%')
ac_versicolor = np.round((j/versicolor_num)*100, 1)
print('K-means versicolor accuracy :{}'.format(ac_versicolor), '%')
ac_virginica = np.round((k/virginica_num)*100, 1)
print('K-means virginica accuracy :{}'.format(ac_virginica), '%')
```
